```python
logger.info(f"Step 3 - Create table {TABLE_NAME_PP01}")
sql = f"""
IF OBJECT_ID('{SCHEMA_NAME}.{TABLE_NAME_PP01}', 'U') IS NOT NULL
    DROP TABLE {SCHEMA_NAME}.{TABLE_NAME_PP01};

CREATE TABLE {SCHEMA_NAME}.{TABLE_NAME_PP01}
(
    transaction_id              NVARCHAR(100),
    transaction_event_code      NVARCHAR(50),
    transaction_initiation_date NVARCHAR(50),
    transaction_updated_date    NVARCHAR(50),
    transaction_amount          NVARCHAR(100),
    transaction_status          NVARCHAR(10),
    transaction_subject         NVARCHAR(MAX),
    ending_balance              NVARCHAR(100),
    available_balance           NVARCHAR(100),
    protection_eligibility      TINYINT,
    payer_info                  NVARCHAR(MAX),
    shipping_info               NVARCHAR(MAX),
    cart_info                   NVARCHAR(MAX),
    store_info                  NVARCHAR(MAX),
    auction_info                NVARCHAR(MAX),
    incentive_info              NVARCHAR(MAX),
    elton_created_at            NVARCHAR(50),
    dt                          NVARCHAR(20),
    hour                        TINYINT
)
"""
create_table(database_name=DATABASE_NAME, sql_server=SQL_SERVER, schema_name=SCHEMA_NAME,
             table_name=TABLE_NAME_PP01, sql=sql, logger=logger)
```

```python
sql2 = f"""
IF OBJECT_ID('{SCHEMA_NAME}.{TABLE_NAME_PP02}', 'U') IS NOT NULL
    DROP TABLE {SCHEMA_NAME}.{TABLE_NAME_PP02};

CREATE TABLE {SCHEMA_NAME}.{TABLE_NAME_PP02}
(
    dispute_id                  NVARCHAR(30),
    create_time                 NVARCHAR(40),
    update_time                 NVARCHAR(40),
    disputed_transactions       NVARCHAR(MAX),
    reason                      NVARCHAR(100),
    status                      NVARCHAR(50),
    dispute_amount              NVARCHAR(100),
    dispute_state               NVARCHAR(30),
    dispute_outcome             NVARCHAR(200),
    adjudications               NVARCHAR(MAX),
    money_movements             NVARCHAR(MAX),
    dispute_life_cycle_stage    NVARCHAR(30),
    dispute_channel             NVARCHAR(20),
    extensions                  NVARCHAR(MAX),
    evidences                   NVARCHAR(MAX),
    seller_response_due_date    NVARCHAR(40),
    offer                       NVARCHAR(MAX),
    refund_details              NVARCHAR(MAX),
    allowed_response_options    NVARCHAR(MAX),
    elton_created_at            NVARCHAR(40),
    dt                          NVARCHAR(20),
    hour                        TINYINT
)
"""
create_table(database_name=DATABASE_NAME, sql_server=SQL_SERVER, schema_name=SCHEMA_NAME, 
             table_name=TABLE_NAME_PP02, sql=sql2, logger=logger)
```
